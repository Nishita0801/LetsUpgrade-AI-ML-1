### converting float values to int

```python
df=pd.DataFrame(s1)
df.CCAvg = df.CCAvg.astype(int)
```

### Applying Logistic Regression

```python
Y=s1['Personal Loan']
X=s1[[ 'Age', 'Experience', 'Income', 'Family', 'CCAvg',
       'Education', 'Mortgage', 'Securities Account',
       'CD Account', 'Online', 'CreditCard']]
import statsmodels.api as sm
x1=sm.add_constant(X)
Logistic=sm.Logit(Y,x1)
result=Logistic.fit()
result.summary()
```

# inferences

##### p value of  attribute Income, Family, CCAvg, Education, Securities Account, CD Account, Online, CreditCard is less than 0.05 
##### so Income, Family, CCAvg, Education, Securities Account, CD Account, Online, CreditCard are highly significant for the sanction of loan
